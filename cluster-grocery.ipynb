## Data Preprocessing
Build item-user matrix

```python
# Convert IDs into counts

def item_count(df):
    item_ids = df['id'].str.split(',').sum() # split the id columns
    return pd.Series(Counter(int(id) for id in item_ids))
```

```python
user_item_counts = history.groupby("user_id").apply(item_count).unstack(fill_value=0)
```

```python
user_item_counts.head()

#each row corresponds to a user
#each column represents an item
#user_item_counts[i,j] shows the quantity of item j bough by user i
```

## Question 1

a. The customer who bought the most items overall in her life

```python
user_total_item = user_item_counts.sum(axis=1)
user_row = user_total_item.argmax()
user_count = user_total_item.max()
user_total_item = user_item_counts.sum(axis=1).reset_index()
user_id= user_total_item.loc[user_row]['user_id']


print("User "+ str(user_id)+ " bought the most items. She bought "+ str(user_count)+" items.")
```

b. For each item, show the customer who bought the item most

```python
item_user_most = user_item_counts.apply(np.argmax, axis=0).reset_index()
item_user_most = item_user_most.rename(columns={'index': 'Item_id',0:'User_id'})
item_user_most.head()
```

```python
df = pd.merge(left = item_user_most, right = item_id, on='Item_id', how='left')
df[['Item_id', 'Item_name', 'User_id']]
```

## Question 2

Item Clustering

```python
# define the feature matrix: item_id, user_id
feature = user_item_counts.T
feature.head()
```

```python
clusters = range(2, 30)
inertias = []
silhouettes = [] # lower is better

for n_clusters in clusters:
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=4)
    kmeans = kmeans.fit(feature)
    label = kmeans.predict(feature)
    
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(feature, label))
```

```python
# visualization
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))
ax[0].plot(clusters, inertias, 'o-', label='Sum of Squared Distances')
ax[0].grid(True)
ax[1].plot(clusters, silhouettes, 'o-', label='Silhouette Coefficient')
ax[1].grid(True)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

The best cluster has the lowest silhouette. Thus, the best number of clusters is 17.

```python
# build K-Means model
kmeans = KMeans(n_clusters=17, init='k-means++', random_state=4)
kmeans = kmeans.fit(feature)
label = kmeans.predict(feature)
```

```python
# PCA for dimension reduction
pca = PCA(n_components=2, random_state=4)
pca = pca.fit(feature)
pca_feature = pca.transform(feature)
```

```python
pca_feature.shape
```

```python
# visualization in 2D

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

fig, ax = plt.subplots(figsize=(16, 16))
for j in range(len(pca_feature)):
    x = pca_feature[j][0]
    y = pca_feature[j][1]
    ax.plot(x, y, 'o', color=colors[label[j] % 8])
    name = item_id[item_id['Item_id'] == j + 1]['Item_name']
    #print(name.values[0][0])
    ax.annotate(xy=(x+0.2, y+0.2), text=name.values[0][:7], fontsize=12)

plt.show()
```
